```python
# order status by age group
df_ecom.groupby(["order_status", "age_group"]).count()
```

```python
pvt_st = pd.pivot_table(data=df_ecom, columns="age_group",
              index=["country", "order_status"],values=["product_name", "category"], aggfunc="count")
```

```python
pvt_st.plot(
    kind='bar',
    stacked=True,
    figsize=(20,6)
)

plt.title('Orders by Country, Status and Age Group')
plt.xlabel('Country and Order Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,8))

sns.heatmap(
    pvt_st.head(10),
    annot=True,
    cmap='YlGnBu'
)
plt.title('Heatmap of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
len(pvt_st)
```

```python
pvt_st.head(15).plot(
    kind='bar',
    figsize=(12,6)
)

plt.title("Orders by Country & Status (Age Group Comparison)")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_reset = pvt_st.reset_index()

pivot_reset.groupby("country").sum().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Country-wise Orders by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
for status in pvt_st.index.get_level_values(1).unique():
    pvt_st.xs(status, level=1).plot(
        kind='bar',
        stacked=True,
        figsize=(20,4),
        title=f"Country-wise Order Status: {status}"
    )
    plt.xticks(rotation=45)
    plt.xlabel("")
    plt.yticks(rotation=45)
    plt.legend()
    plt.show()
```
